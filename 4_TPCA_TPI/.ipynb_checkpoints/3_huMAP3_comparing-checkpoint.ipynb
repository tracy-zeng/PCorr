```python
import pickle
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# read datasets
seeds = [0,42,100,1000,10000,100000,1000000,10000000,100000000,1000000000]
r_humap = pickle.load(open('TPCA_TPI_datasets/r_humap.pkl', 'rb'))
r_randomhumap = pickle.load(open('TPCA_TPI_datasets/r_randomhumap.pkl', 'rb'))
```

```python
r_humap['score'] = np.mean(r_humap, axis=1)
r_randomhumap['score'] = np.mean(r_randomhumap, axis=1)

plt.figure(figsize=(6, 6))
sns.kdeplot(r_randomhumap['score'], cut=0, fill=False, color='#329932',ls='--',
            alpha=1, linewidth=2,label='Random')
sns.kdeplot(r_humap['score'], cut=0, fill=True, color='#329932',
            alpha=0.2, linewidth=2,label='hu.MAP 3.0')
plt.yticks(ha='right', fontsize=22)
plt.xticks(ha='right', fontsize=22, rotation=0)
plt.xlabel('TPCA-TPI score', fontsize=26, labelpad=20)
plt.ylabel('Density', fontsize=26, labelpad=20)
plt.xlim(0, 1)
plt.legend(loc=2, bbox_to_anchor=(1.05, 1.0), fontsize=17, markerfirst=True,
           markerscale=1, scatterpoints=1, ncol=1, title_fontsize=17)
plt.show()
```
