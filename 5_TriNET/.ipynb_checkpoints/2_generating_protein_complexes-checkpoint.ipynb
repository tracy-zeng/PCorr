```python
import pickle
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
# read datasets
tpis_sig_key = pickle.load(open('TriNET_datasets/tpis_sig_key.pkl', 'rb'))
```

```python
pros_key = {}
for ge_key in tpis_sig_key.keys():
    pros = set()
    tpis_sig = tpis_sig_key[ge_key]
    for t in tpis_sig:
        pros.add(t[0])
        pros.add(t[1])
        pros.add(t[2])
    pros_key[ge_key] = pros
```

```python
tpis_pro_key = {}
for ge_key in tpis_sig_key.keys():
    tpis_pro_key[ge_key] = {}
    tpis_sig = tpis_sig_key[ge_key]
    for t in tqdm(tpis_sig):
        try:
            tpis_pro_key[ge_key][t[0]].add(t)
        except:
            tpis_pro_key[ge_key][t[0]] = set()
            tpis_pro_key[ge_key][t[0]].add(t)

        try:
            tpis_pro_key[ge_key][t[1]].add(t)
        except:
            tpis_pro_key[ge_key][t[1]] = set()
            tpis_pro_key[ge_key][t[1]].add(t)

        try:
            tpis_pro_key[ge_key][t[2]].add(t)
        except:
            tpis_pro_key[ge_key][t[2]] = set()
            tpis_pro_key[ge_key][t[2]].add(t)
```
